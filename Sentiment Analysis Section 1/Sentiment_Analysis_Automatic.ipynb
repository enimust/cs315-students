##**Training Model**

```python
import nltk

from nltk import word_tokenize
from nltk.corpus import stopwords

from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
from nltk.stem import LancasterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')

ps = PorterStemmer()
la = LancasterStemmer()

import gensim

import numpy as np
import pandas as pd
import re

import transformers
from transformers import pipeline
```

```python
## This function takes in a list of text where it identifies words of texts that contains extended words.
## The function then replaces extended words into normal words. For example, "hiiiiii" will be returned as "hi".
def rm_extendWords(list):
  outList = []
  comJoin = []

  # print(list)
  for comment in list:
    comment = comment.split()
    # print(comment)
    if len(comment) != 1:
      for w in comment:
        # print(w)
        if len(set(w)) != len(w):
          lastChar = w[-1]
          remove = re.findall('.*?(' + re.escape(lastChar) + '{3,})', w)
          if len(remove) == 1:
            remove = remove[0]
            outW = re.sub('.*?(' + re.escape(lastChar) + '{3,})', w.replace(remove, lastChar), w);
            # print(outW)
            comJoin.append(outW)
          else:
            comJoin.append(w)
        else:
          comJoin.append(w)
      outList.append(' '.join(comJoin))
      comJoin = []
  return outList
```

```python
#remove expanded words like "hiii"
data1 = rm_extendWords(data)
print(data1)
print(data)
```

```python
# add tags to the train sampling

def process(text) :
  #remove non letters
  rawText = re.sub('[^a-zA-Z]', " ", text)

  #remove numbers
  rawText = re.sub('\d+(\.\d+)?', " ", rawText)

  #remove hahas, not convertible to float later on, and has neutral connotation
  rawText = re.sub('(ha)+', " ", rawText)

  #lowercase
  words = rawText.lower().split()

  #define stopwords (e.g. a, the, an, of)
  stop = set(stopwords.words("english"))

  key_words = [ps.stem(w) for w in words if not w in stop]

  return (" ".join(key_words))
```

```python
cleantext = []
for i in range(0, len(data1)) :
  cleantext.append(process(data1[i]))

print(data1)
```

##**Predicting Model**

```python
# create a feature extractor that transfers the text input to a feature vector --> ... change

from sklearn.model_selection import train_test_split

train_text, test_text, train_label, test_label = train_test_split(df['CleanTexts'], df['Sentiment'], test_size=0.2)
```

```python
vectorizer = CountVectorizer(
    strip_accents='unicode',
    stop_words='english',
    lowercase=True,
    token_pattern=r'\b[a-zA-Z]{3,}\b', # we want only words that contain letters and are 3 or more characters long
)

# Transform our data into the document-term matrix
train_feature = vectorizer.fit_transform(train_text)
test_feature = vectorizer.transform(test_text)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report

svm_model = SVC(kernel="linear")
svm_model.fit(train_feature, train_label)
```

```python
pred_svm = svm_model.predict(test_feature)
```

```python
print(classification_report(test_label, pred_svm))
```

```python
from sklearn.naive_bayes import GaussianNB

bayes_model = GaussianNB()
bayes_model.fit(train_feature.toarray(), train_label)
```

```python
pred_bayes = bayes_model.predict(test_feature.toarray())
```

```python
print(classification_report(test_label, pred_bayes))
```
