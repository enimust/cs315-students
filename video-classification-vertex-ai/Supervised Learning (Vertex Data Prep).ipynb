## Making the dataframe using data directories

For this to work, you want your videos to be organized in separate directories for ads and nonads. For example, we have our directory name ```{label}_vids_{userID}```. In each directory are mp4 files collected using the Pyktok library.

````python
def makeDF(vidtype):
    """
    given a label, make a dataframe using a directory of videos
    with filename and the label
    """
    df = pd.DataFrame(os.listdir(f'{vidtype}_vids_33534'),columns=['file']) #change the f-string to fit your directory
    df['label'] = vidtype
    return df
````

````python
# make a dataframe
ads_df = makeDF("ads")
nonads_df = makeDF("nonads")
````

````python
# create one big dataframe for all videos
labels = pd.concat([ads_df,nonads_df])

# make a new index
labels = labels.reset_index()
labels = labels.drop(['index'],axis=1)
labels
````

## Make an import file

Make an import file for video classification

````python
def uriCol(name):
    """
    finds the URI that links to cloud storage location
    of one video using the file name
    """
    return f"gs://your_uri_path/{name}"

labels['video_uri'] = labels['file'].apply(uriCol)
labels['start'] = '0'
labels['end'] = 'inf'
````

````python
cols = ['video_uri','label','start','end']
labels = labels[cols]
labels
````

````python
# remove column headers and column of indices to avoid confusing the model
labels.to_csv('vid_import_file.csv',index=False,header=False)
````
