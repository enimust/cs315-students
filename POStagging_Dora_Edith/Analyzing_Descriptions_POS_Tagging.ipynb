```python
def save_spacy_doc(df):
    df_final = df.copy()
    df_final['spacy_doc'] = df_final['description'].apply(lambda desc: nlp(desc))
    return df_final

def get_pos_list(doc, pos):
    pos_list = [token for token in doc if token.pos_ == pos]
    return pos_list

def add_pos_list_columns(df):
    df_final = df.copy()
    df_final['nouns'] = df_final['spacy_doc'].apply(lambda doc: get_pos_list(doc, 'NOUN'))
    df_final['prop_nouns'] = df_final['spacy_doc'].apply(lambda doc: get_pos_list(doc,'PROPN'))
    df_final['adjs'] = df_final['spacy_doc'].apply(lambda doc: get_pos_list(doc, 'ADJ'))
    df_final['advs'] = df_final['spacy_doc'].apply(lambda doc: get_pos_list(doc, 'ADV'))
    return df_final

# def get_ent_label_list(doc):
#     ent_label_list = [(ent, ent.label_) for ent in doc.ents]
#     return ent_label_list

def get_entities(doc):
    ents = [ent for ent in doc.ents]
    return ents

def get_entity_labels(doc):
    labels = [ent.label_ for ent in doc.ents]
    return labels

def add_entity_and_label_columns(df):
    df_final = df.copy()
    df_final['entities'] = df['spacy_doc'].apply(get_entities)
    df_final['entity_labels'] = df['spacy_doc'].apply(get_entity_labels)
    return df_final

def make_lowercase(word_list):
    if len(word_list)>0:
        lowercase_list = [str(word).lower() for word in word_list]
    else:
        lowercase_list = []
    return lowercase_list

def make_lowercase_and_remove_punct(word_list):
    clean_list = 
    for word in word_list:
        "".join([char for char in word if char not in string.punctuation])

def add_lowercase_pos_columns(df):
    df_final = df.copy()
    df_final['clean_nouns'] = df_final['nouns'].apply(make_lowercase)
    df_final['clean_prop_nouns'] = df_final['prop_nouns'].apply(make_lowercase)
    df_final['clean_adjs'] = df_final['adjs'].apply(make_lowercase)
    df_final['clean_advs'] = df_final['advs'].apply(make_lowercase)
    df_final['clean_ents'] = df_final['advs'].apply(make_lowercase)
    return df_final
```
